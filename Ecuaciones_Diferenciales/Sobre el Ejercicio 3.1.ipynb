# Péndulo amortiguado

En un péndulo con varilla de masa despreciable de longitud $L$ y masa puntual $m$ en el extremo, el modelo que describe el movimiento es
$$
\ddot\theta=-\frac{g}{L}\sin\theta-\frac{b}{mL}\dot\theta,
$$
donde $b$ es el factor de fricción viscosa.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

Reescribiendo las variables como $x=[x_1,x_2]^T=[\theta,\dot\theta]^T$, se tiene
$$
\dot x=\begin{bmatrix}x_2\\
-\frac{g}{L}\sin(x_1)-\frac{b}{mL}x_2.
\end{bmatrix}
$$

```python
g=9.8
def pendulo(x,t,m=1,L=1,b=0.5):
    # Inizialización de dx
    dx=np.zeros(2) # Regresa un arreglo de 2 ceros
    dx[0]=x[1]
    dx[1]=-g/L*np.sin(x[0])-b/(m*L)*x[1]
    return dx
```

```python
t=np.linspace(0,10,100)
x0=[np.pi/2,0]
x=odeint(pendulo,x0,t)
x.shape
```

```python
plt.plot(t,x[:,0],label='Posición')
plt.plot(t,x[:,1],label='Velocidad')
plt.legend()
plt.grid()
```
